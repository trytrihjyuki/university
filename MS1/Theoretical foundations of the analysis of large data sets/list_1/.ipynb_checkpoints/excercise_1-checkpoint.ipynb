```python
from scipy.stats import rv_continuous
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pylab as py
from scipy.stats import norm

class distribution(rv_continuous):
    def _pdf(self, x, alpha):
        return (alpha + 1) * (x ** alpha)

dst = distribution(a=0, b=1, name="custom dst")
```

```python
def make_experiment(num_of_times, n, alpha):
    mle_est = []
    moment_est = []

    for i in range(num_of_times):
        sample = dst.rvs(alpha=alpha, size=n)
        log_sum = sum(np.log(sample))
        sample_mean = sample.mean()

        mle_est.append(- (n + log_sum) / log_sum)
        
        moment_est.append((2*sample_mean - 1) / (1 - sample_mean))

    return mle_est, moment_est
```

```python
mle_est, moment_est = make_experiment(1000, 20, 5)
```

```python
plt.hist(x=mle_est, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
plt.show()

plt.hist(x=moment_est, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
```

```python
plot = plt.boxplot(mle_est)
plt.show()
plot = plt.boxplot(moment_est)
```

```python
sm.qqplot(np.array(mle_est), line ='45')
py.show()
sm.qqplot(np.array(moment_est), line ='45')
py.show()
```

```python
alpha = 5
n = 20
results = {"var" : [], "MSE" : [], "bias" : [], "avg est value" : [], "quantiles": []}
for est in [np.array(mle_est), np.array(moment_est)]:
    var = np.var(est)
    bias = sum(est - alpha)/est.shape[0]
    mse = bias * bias + var 
    avg_value = est.mean()
    
    results["var"].append(var)
    results["MSE"].append(mse)
    results["bias"].append(bias)      
    results["avg est value"].append(avg_value)

    sorted_est = sorted(est)
    quantile_05 = sorted_est[25 - 1]
    quantile_95 = sorted_est[975 - 1]

    results["quantiles"].append((quantile_05, quantile_95))

print(f"Asymptotic normal distribution of MLE:\n")
print(f"N({alpha},{(alpha+1)**2 / n:.2f}) VAR: {(alpha+1)**2 / n} MSE: {(alpha+1)**2 / n} BIAS: {0.0}")
print(f"Q_05 : {norm.ppf(0.025, loc=alpha, scale=(alpha+1)**2 / n)} Q_95 : {norm.ppf(0.975, loc=alpha, scale=(alpha+1)**2 / n)}")

pd.DataFrame(results, index=["MLE est", "Moment est"])
```
